```python

def show_history():
    df_sheet_main = pd.read_csv('/opt/app/notebook/sheet1.csv',sep=",",low_memory=False)
    df_sheet = df_sheet_main
    
    
    yrs=df_sheet['year'].to_numpy()
    drama=df_sheet['Drama'].to_numpy()
    romance=df_sheet['Romance'].to_numpy()
    action=df_sheet['Action'].to_numpy()
    adventure=df_sheet['Adventure'].to_numpy()
    comedy=df_sheet['Comedy'].to_numpy()
    war=df_sheet['War'].to_numpy()
    scfif=df_sheet['Sci-Fi'].to_numpy()
    History=df_sheet['History'].to_numpy()
    Documentary=df_sheet['Documentary'].to_numpy()
    Mystery=df_sheet['Mystery'].to_numpy()
    Horror=df_sheet['Horror'].to_numpy()
    Musical=df_sheet['Musical'].to_numpy()
    News=df_sheet['News'].to_numpy()
    plt.figure(figsize=(15,5),dpi=300)   # This command is optional

    # Plot y versus x as lines and/or markers
    plt.plot( yrs, drama, '--' ,label="Drama",)
    plt.plot( yrs, romance, '--' ,label="Romance",)
    plt.plot( yrs, action, '--' ,label="Action",)
    plt.plot( yrs, adventure, '--',label="Adventure", )
    plt.plot( yrs, comedy, '--' ,label="Comedy",)
    plt.plot( yrs, war, '--' ,label="War",)
    plt.plot( yrs, scfif, '-' ,label="Sci-Fi",)
    plt.plot( yrs, Horror, '--' ,label="Horror",)
    plt.plot( yrs, Musical, '--' ,label="Musical",)
    plt.plot( yrs, News, '--' ,label="News",)

    plt.annotate("WW I", 
             xy=(1919, 400), xycoords='data', 
             xytext=(50,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    plt.annotate("TV", 
             xy=(1928, 400), xycoords='data', 
             xytext=(50,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    plt.annotate("WW II", 
             xy=(1940, 400), xycoords='data', 
             xytext=(50,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    plt.annotate("Cold War", 
             xy=(1941, 400), xycoords='data', 
             xytext=(50,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    plt.annotate("End Cold War", 
             xy=(1991, 400), xycoords='data', 
             xytext=(50,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2'))



    plt.annotate("Project Apollo", 
             xy=(1961, 400), xycoords='data', 
             xytext=(50,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))

    plt.annotate("Apollo 11", 
             xy=(1969, 1000), xycoords='data', 
             xytext=(50,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    plt.annotate("9/11", 
             xy=(2001, 2000), xycoords='data', 
             xytext=(50,90), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.5'))

    plt.annotate("Covid19", 
             xy=(2019, 2500), xycoords='data', 
             xytext=(55,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))
    plt.title(f"History")
    plt.xlabel("Year")
    plt.ylabel("NumOfMovies")
    
    plt.legend()
    
def show_history_scifi():
    df_sheet_main = pd.read_csv('/opt/app/notebook/sheet1.csv',sep=",",low_memory=False)
    df_sheet = df_sheet_main
    
    
    yrs_end=1930
    yrs=df_sheet.loc[df_sheet['year'] <yrs_end,['year']].to_numpy()

    drama=df_sheet.loc[df_sheet['year'] <yrs_end,['Drama']].to_numpy()
    romance=df_sheet.loc[df_sheet['year'] <yrs_end,['Romance']].to_numpy()
    action=df_sheet.loc[df_sheet['year'] <yrs_end,['Action']].to_numpy()
    adventure=df_sheet.loc[df_sheet['year'] <yrs_end,['Adventure']].to_numpy()
    comedy=df_sheet.loc[df_sheet['year'] <yrs_end,['Comedy']].to_numpy()
    war=df_sheet.loc[df_sheet['year'] <yrs_end,['War']].to_numpy()
    scfif=df_sheet.loc[df_sheet['year'] <yrs_end,['Sci-Fi']].to_numpy()
    fantasy=df_sheet.loc[df_sheet['year'] <yrs_end,['Fantasy']].to_numpy()
    
    plt.figure(figsize=(15,5),dpi=300)   # This command is optional
    
    plt.annotate("WW I (1919)",
    xy=(1919, 1030), xycoords='data', 
    xytext=(50,0), textcoords='offset points', arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2'))

    # Plot y versus x as lines and/or markers
    plt.plot( yrs, drama, '--' ,label="Drama",)
    plt.plot( yrs, action, '--' ,label="Action",)
    plt.plot( yrs, adventure, '--',label="Adventure", )
    plt.plot( yrs, comedy, '--' ,label="Comedy",)
    plt.plot( yrs, war, '--' ,label="War",)
    plt.plot( yrs, scfif, '-' ,label="Sci-Fi",)

    plt.title(f"History")
    plt.xlabel("Year")
    plt.ylabel("NumOfMovies")
    
    plt.legend()
    
    

def gross_scifi_and_fantasy(df_full_sheet,startYear,endYear):
    
    gross_sum_by_yrs = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear)].groupby('startYear')["lifetime_gross"].sum()
    
    gross_sci_without_fan = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  ~df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    gross_sciadv_without_fan = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Adventure",case=False,na=False) \
                & df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  ~df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    gross_scidrama_without_fan = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Drama",case=False,na=False) \
                & df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  ~df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    # Fantasy
    gross_fan_without_sci = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & ~df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    gross_fanadv_without_sci = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Adventure",case=False,na=False) \
                & ~df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    gross_fandrama_without_sci = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Drama",case=False,na=False) \
                & ~df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

# Fantasy Sci-Fi
    gross_fan_and_sci = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                & df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()

    gross_drama_pure = df_full_sheet.loc[(df_full_sheet['startYear'] >= startYear) & (df_full_sheet['startYear'] < endYear) \
                 & ~df_full_sheet['genres'].str.contains("Drama",case=False,na=False)
                & ~df_full_sheet['genres'].str.contains("Sci-Fi",case=False,na=False) \
                &  ~df_full_sheet['genres'].str.contains("Fantasy",case=False,na=False)] \
                .groupby('startYear')["lifetime_gross"].sum()
    
    
    # sci-fi
    df_full_sheet2 = pd.merge(gross_sum_by_yrs.to_frame().rename(columns={'lifetime_gross':'gross_all'}) \
                         ,gross_sci_without_fan.to_frame().rename(columns={'lifetime_gross':'gross_sci'}) \
                         , how='outer', left_on = 'startYear', right_on = 'startYear')

    df_full_sheet2 = pd.merge(df_full_sheet2 \
                         ,gross_sciadv_without_fan.to_frame().rename(columns={'lifetime_gross':'gross_sciadv_without_fan'}) \
                         , how='outer', left_on = 'startYear', right_on = 'startYear').reset_index().replace(np.nan, 0)


    # fantasy
    df_full_sheet2 = pd.merge(df_full_sheet2 \
                          ,gross_fan_without_sci.to_frame().rename(columns={'lifetime_gross':'gross_fan'}) \
                          , how='outer', left_on = 'startYear', right_on = 'startYear').reset_index().replace(np.nan, 0)
    df_full_sheet2 = pd.merge(df_full_sheet2 \
                         ,gross_fanadv_without_sci.to_frame().rename(columns={'lifetime_gross':'gross_fanadv_without_sci'}) \
                         , how='outer', left_on = 'startYear', right_on = 'startYear').reset_index().replace(np.nan, 0)
    
    df_full_sheet2['gross_sci_per'] = df_full_sheet2[['gross_all','gross_sci']].apply(lambda x: x['gross_sci']*100/x['gross_all'], axis=1)
    df_full_sheet2['gross_fan_per'] = df_full_sheet2[['gross_all','gross_fan']].apply(lambda x: x['gross_fan']*100/x['gross_all'], axis=1)
    df_full_sheet2['gross_sciadv_without_fan_per'] = df_full_sheet2[['gross_all','gross_sciadv_without_fan']].apply(lambda x: x['gross_sciadv_without_fan']*100/x['gross_all'], axis=1)
    df_full_sheet2['gross_fanadv_without_sci_per'] = df_full_sheet2[['gross_all','gross_fanadv_without_sci']].apply(lambda x: x['gross_fanadv_without_sci']*100/x['gross_all'], axis=1)
    
    
    return(df_full_sheet2)
    

def display_gross_scifi_and_fantasy(df_full):
    
    plt.style.use('default')
    plt.figure(figsize=(15,10),dpi=300)
    plt.plot( df_full['startYear'], df_full['gross_sci_per'], '-' ,label="SCI-FI",)
    plt.plot( df_full['startYear'], df_full['gross_fan_per'], '-' ,label="Fantasy",)


    plt.annotate("Star Wars: Episode IV - A New Hope (1977)", 
             xy=(1977, 29), xycoords='data', 
             xytext=(5,100), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.2'))


    plt.annotate("Star Wars: Episode V \n- The Empire Strikes Back (1980)", 
             xy=(1980, 16), xycoords='data', 
             xytext=(60,105), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))

    plt.annotate("Star Wars: Episode VI - Return of the Jedi (1983)", 
             xy=(1983, 18), xycoords='data', 
             xytext=(50,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))


    plt.annotate("2001: A Space Odyssey\n & Planet of the Apes (1968)", 
             xy=(1968, 20), xycoords='data', 
             xytext=(-5,140), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))

    plt.annotate("A Clockwork \nOrange (1971)", 
             xy=(1971, 12), xycoords='data', 
             xytext=(-10,120), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))

    plt.annotate("Star Trek: The Motion Picture (1979)", 
             xy=(1979, 20), xycoords='data', 
             xytext=(10,130), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))

    plt.annotate("E.T. (1982)", 
             xy=(1982, 24), xycoords='data', 
             xytext=(50,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1'))
    
    plt.annotate("Star Wars: The Force Awakens (2015)", 
             xy=(2015, 38), xycoords='data', 
             xytext=(-110,50), textcoords='offset points', 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=2'))
    
    plt.xlabel("Year")
    plt.ylabel("percentage")


    plt.legend()


    
```

```python
def showbar_topTwenty(df,genres,startYear,endYear):
    scifi_yrs_list = df.loc[(df['startYear'] >= startYear) & (df['startYear'] < endYear)\
                            & (df['genres'].str.contains(genres,case=False,na=False)),:]['startYear'].unique().tolist()

    cc=0
    for i in scifi_yrs_list:
        #print(i)
        #df.loc[(df['genres'].str.contains("Sci-Fi",case=False,na=False)) & (df['startYear'] == i),:]
        df_buf = df.loc[df['startYear'] == i,['primaryTitle','genres','lifetime_gross']] \
            .sort_values(by=['lifetime_gross'], ascending=False).iloc[0:20].reset_index(drop=True)
    
        numOfsci = df_buf.loc[(df_buf['genres'].str.contains(genres,case=False,na=False)),:]['primaryTitle'].count()
        data = [[i, numOfsci]]
        df_box = pd.DataFrame(data, columns=['yrs', 'num'])

        if cc == 0:
            dfff = df_box
        else:
            dfff = pd.concat([dfff, df_box])
        
        cc += 1

    dfff = dfff.sort_values(by=['yrs'], ascending=True).reset_index(drop=True)
    
    plt.style.use('fivethirtyeight')
    
    plt.figure(figsize=(15,5),dpi=300)
    #plt.plot( dfff['yrs'], dfff['num'], '-' ,label=genres,)
    plt.bar(dfff['yrs'],dfff['num'],label="TT")
    plt.title(genres)
    plt.xlabel("Year")
    plt.ylabel("Count")


def showbar_topTwenty_except(df,genres,genres_except,startYear,endYear):
    scifi_yrs_list = df.loc[(df['startYear'] >= startYear) & (df['startYear'] < endYear) \
                            & (df['genres'].str.contains(genres,case=False,na=False)),:]['startYear'].unique().tolist()

    cc=0
    for i in scifi_yrs_list:
        #print(i)
        #df.loc[(df['genres'].str.contains("Sci-Fi",case=False,na=False)) & (df['startYear'] == i),:]
        df_buf = df.loc[df['startYear'] == i,['primaryTitle','genres','lifetime_gross']] \
            .sort_values(by=['lifetime_gross'], ascending=False).iloc[0:20].reset_index(drop=True)
    
        numOfsci = df_buf.loc[(df_buf['genres'].str.contains(genres,case=False,na=False)) \
                              & ~(df_buf['genres'].str.contains(genres_except,case=False,na=False)),:]['primaryTitle'].count()
        data = [[i, numOfsci]]
        df_box = pd.DataFrame(data, columns=['yrs', 'num'])

        if cc == 0:
            dfff = df_box
        else:
            dfff = pd.concat([dfff, df_box])
        
        cc += 1

    dfff = dfff.sort_values(by=['yrs'], ascending=True).reset_index(drop=True)
    
    plt.style.use('fivethirtyeight')
    
    plt.figure(figsize=(15,5),dpi=300)
    #plt.plot( dfff['yrs'], dfff['num'], '-' ,label=genres,)
    plt.bar(dfff['yrs'],dfff['num'])
    plt.title(f"{genres} without {genres_except}")
    plt.xlabel("Year")
    plt.ylabel("Count")

def currency(x, pos):
    """The two arguments are the value and tick position"""
    if x >= 1e8:
        s = '${:1.1f}M'.format(x*1e-8)
    else:
        s = '${:1.0f}K'.format(x*1e-8)
    return s

def showbar_TopTen_by_genres(df,startYear,endYear):

    df_top = df.loc[(df['startYear'] >= startYear) & (df['startYear'] < endYear) \
                    ,['tconst','primaryTitle','startYear','genres','lifetime_gross']] \
            .sort_values(by=['lifetime_gross'], ascending=False) \
            .iloc[0:10].reset_index(drop=True)
    
    ind = np.arange(10)
    
    list_name = df_top['primaryTitle'].tolist()
    list_genres = df_top['genres'].tolist()
    y_pos = np.arange(len(list_name))
    
    plt.rcdefaults()
    fig, ax = plt.subplots()
    ax.barh(y_pos, df_top['lifetime_gross'], align='center')
    ax.set_yticks(y_pos, labels=list_name)
    ax.xaxis.set_major_formatter(currency)
    #ax.set_yticks(ind, labels=list_genres)

    ax.set_xlabel('lifetime gross')
    ax.invert_yaxis()
    text=f"{startYear} - {endYear}"
    ax.set_title(text)
    #ax.legend()

    plt.show()
    
    #return(df_top)
    
```
